```python
import numpy as np
from math import factorial
```

```python
def Binom(n, k, p):
    b = factorial(n) // (factorial(k) * factorial(n - k))
    return b * p ** k * (1 - p) ** (n - k)
```

```python
y = np.array([7, 7, 1, 3, 2, 1, 5, 1, 4, 0, 2, 3, 3, 1, 2, 2, 3, 1, 4, 0, 8, 3, 5, 3, 0, 0, 7, 1, 1, 3, 1, 3, 2, 4, 1, 6, 2, 2, 4, 1, 3, 1, 1, 2, 7, 3, 3, 2, 2, 2])
M = 10
eps = 1e-6
```

```python
x = np.random.rand()
pi_0 = np.array([x, 1 - x])
mu_0 = np.random.rand(2)

pi_1 = np.array([331, 331])
mu_1 = np.array([331, 331])
```

```python
iteration = 0

while np.abs(pi_0 - pi_1).sum() + np.abs(mu_0 - mu_1).sum() > eps:
    print('iteration = ' + str(iteration), 'pi = ' + str(pi_0), 'mu = ' + str(mu_0), sep='\n', end='\n\n')
    pi_1, mu_1 = pi_0, mu_0
    iteration += 1
    
    r = []
    for k in y:
        a = []
        for p in mu_0:
            a.append(Binom(M, k, p))
        r.append(a)
    r = np.array(r) 
    r *= pi_0[None,:]
    r /= np.sum(r, axis = 1)[:, None]

    pi_0 = np.average(r, axis = 0)
    mu_0 = np.sum(y[:, None] * r, axis = 0) / (r.sum(axis = 0) * M)
    
print('\nanswer:', 'pi = ' + str(pi_0), 'mu = ' + str(mu_0), sep='\n')
```
